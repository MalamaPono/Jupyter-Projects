```python
import PIL
from PIL import Image
from IPython.display import display
import numpy
import pandas

file = 'data/msi_recruitment.gif'
image =Image.open(file)
image = image.convert('RGB')
```

```python
#to draw the black bar with the text
from PIL import ImageDraw
from PIL import ImageFont

font = ImageFont.truetype("data/fanwood-webfont.ttf", 28)

images = []
intensities = [.1,.5,.9]
#.1,.5,.9


drawing_object = ImageDraw.Draw(image)

for i in range(3):
    for j in range(3):
        drawing_object.rectangle((0, image.height-35, image.width, image.height),fill='black')
        drawing_object.text((10, image.height-30),"channel " + str(i) + " intensity " + str(intensities[j]),(255,255,255),font=font)
        colors = image.split()
        colors = list(colors)
        colors[i] = colors[i].point(lambda x : x*intensities[j])
        result = Image.merge('RGB', tuple(colors))
        images.append(result)
```

```python
first_image = images[0]
contact_sheet=PIL.Image.new(first_image.mode, (first_image.width*3,first_image.height*3))

x=0
y=0

for image in images:
    contact_sheet.paste(image,(x,y))
    if x+first_image.width == contact_sheet.width:
        x = 0
        y+=first_image.height
    else:
        x+=first_image.width

contact_sheet = contact_sheet.resize((int(contact_sheet.width/2),int(contact_sheet.height/2)))
display(contact_sheet)
```
