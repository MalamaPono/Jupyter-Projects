```python
import PIL 
from PIL import Image
from IPython.display import display

file = 'data/msi_recruitment.gif'
image = Image.open(file)
```

```python
help(image.save)
```

```python
image.save('msi_recuitment.png')
image = Image.open('msi_recuitment.png')
image.show()
```

```python
from PIL import ImageFilter
help(ImageFilter)
```

```python
image = image.convert('RGB')
blurred_image = image.filter(PIL.ImageFilter.BLUR)
display(blurred_image)
```

```python
sharpened_image = image.filter(PIL.ImageFilter.SHARPEN)
display(sharpened_image)
```

```python
print('{}x{}'.format(image.width,image.height))
```

```python
help(image.crop)
```

```python
#left upper vertex coordinate,right lower vertex coordinate to make a rectangle that you want to crop out of the image. 
#usually pixels start at 0,0 in the top left corner and pixels are named by (column_number, row_number)
# going to the right, the x coordinate increases, while going down the y-coordinate increases

michigan_logo = image.crop((50,0,190,150))
display(michigan_logo)
```

```python
from PIL import ImageDraw

drawing_object = ImageDraw.Draw(image)
drawing_object.rectangle((50,0,190,150),fill = None, outline='red')
display(image)
```
