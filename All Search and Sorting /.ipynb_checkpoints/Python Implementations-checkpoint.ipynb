```python
# quicksort

def quicksort(num,left,right):
    
    if(left >= right):
        return
    
    pivotValue = num[(left+right)/2]
    partitionIndex = partition(num, left, right, pivotValue)     
    quickSort(num,left,partitionIndex-1)
    quickSort(num,partitionIndex,right)
    
def partition(num,left,right,pivotValue)
    
    while(left <= right):
        while(num[left] < pivotValue)
            left++
                
        while(num[right] > pivotValue)
            right--
                
        if(left <= right)
            temp = num[left]
            num[left] = num[right]
            num[right] = temp
            right-=1
            left+=1
        
    return left
```

```python
# mergesort

def merge(arr,left,mid,right):
    temp = []
    
    left_pointer = left
    left_end = mid
    
    right_pointer = mid + 1
    right_end = right
    
    while (left_pointer <= left_end) and (right_pointer <= right_end):
        if arr[left_pointer] < arr[right_pointer]:
            temp.append(arr[left_pointer])
            left_pointer+=1
        else:
            temp.append(arr[right_pointer])
            right_pointer+=1
            
    while left_pointer <= left_end:
        temp.append(arr[left_pointer])
        left_pointer+=1
        
    while right_pointer <= right_end:
        temp.append(arr[right_pointer])
        right_pointer+=1
    
    
    # copy back from temp storage into the original arr
    idx = 0
    for i in range(left,right+1):
        arr[i] = temp[idx]
        idx+=1
 
def mergesort(arr,left,right):
    if left >= right:
        return
    
    elif left + 1 == right:
        if arr[right] < arr[left]:
            temp = arr[left]
            arr[left] = arr[right]
            arr[right] = temp
        return
    
    else:
        mid = (left + right) // 2
        mergesort(arr,left,mid)
        mergesort(arr,mid+1,right)
        merge(arr,left,mid,right)
        
arr = [-5,2,3,-98,5,6]
mergesort(arr,0,len(arr) - 1)
print(arr)
```

```python
# binary search
def binarySearchRecursive(num,low,high,target)
    mid = (low + high)/2
    
    if(low <= high):
        if(target > num[mid]):
            return binarySearchRecursive(num, mid+1, high, target)
        elif(target < num[mid]):
            return binarySearchRecursive(num, low, mid-1, target)
        else:
            return mid

    else:
        return -1
            
        
    
def binarySearchIterative(num, target)
    low = 0
    high = num.length-1
    while(low <= high):
        mid = (low+high)/2
        if(target > num[mid]):
            low = mid+1
        elif(target < num[mid]):
            high = mid -1
        else:
            return mid


    return -1    
```

```python
# linear search

def linearSearch(arr,target)
    for i in range(arr):
        if arr[i] == target:
            return i
```
